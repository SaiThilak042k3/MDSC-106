# **Line plot Example**

>   

*   *In this example, x is created using numpy library linespace from 0 to 2, In between 200 points.*

*   *Finally, the graph is plotted using the plot() method of matplotlib.pyplot.*

*   *Next, the X and Y axis are labeled and the graph is given a title.*

*   *Here the line style is used. The line style chosen is ‘--‘,‘Dotted‘,‘Dashdot‘.*

*   *Legend is Add at the end for better understanding.*

```python
x = np.linspace(0, 2, 100)  # Line plot.


#plt.subplots(figsize=(5, 2.7), layout='constrained')
plt.plot(x, x, label='linear',linestyle='--')  # Plot some data on the axes.
plt.plot(x, x**2, label='quadratic',linestyle='dotted')  # Plot more data on the axes...
plt.plot(x, x**3, label='cubic',linestyle='dashdot')  # ... and some more.
plt.xlabel('x label')  # Add an x-label to the axes.
plt.ylabel('y label')  # Add a y-label to the axes.
plt.title("Simple Plot")  # Add a title to the axes.
plt.legend()  # Add
```

# **Histogram**

*A histogram is a graphical representation of the distribution of data given by the user. Its appearance is similar to Bar-Graph except it is continuous.*

*The towers or bars of a histogram are called bins. The height of each bin shows how many values from that data fall into that range.* 

```python
sns.histplot(x=df['math score'],hue=df['gender'],bins=10, kde=True)

# from this histogram girls students scored more than boys in math 
```

```python
sns.histplot(df, x="race/ethnicity", hue="gender", element="step")
# in the plot we can say that group c female are more  and in group b male are more 
```

Overlapping bars can be hard to visually resolve. A different approach would be to draw a step function:

You can move even farther away from bars by drawing a polygon with vertices in the center of each bin. This may make it easier to see the shape of the distribution, but use with caution: it will be less obvious to your audience that they are looking at a histogram:

```python
sns.histplot(df, y="parental level of education", hue="gender", element="poly")
# by this graph we can clearly see that most of the parents did some college level of eductaion
```

```python


# Generating  attendance data
np.random.seed(0)
months = ['January', 'February', 'March', 'April']
classes = ['Class A', 'Class B', 'Class C', 'Class D']

data = {
    'Month': [month for _ in range(len(classes)) for month in months],
    'Class': [cls for cls in classes for _ in range(len(months))],
    'Attendance': np.random.randint(50, 100, size=len(months) * len(classes))
}

df = pd.DataFrame(data)

# Set the style of the plot
sns.set(style="whitegrid")

# Create a stacked bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Month', y='Attendance', hue='Class', data=df, palette='bright')

# Customize the plot
plt.xlabel('Month')
plt.ylabel('Attendance Count')
plt.title('Monthly Student Attendance by Class')
plt.legend(title='Class', loc='upper right')

# Show the plot
plt.xticks(rotation=45)
plt.show()
```

We have attendance data for four different classes (Class A, Class B, Class C, and Class D) over four months (January, February, March, and April).
