# perform below mentioned dataframe operations on the data:

    - add 2 synthetic columns with column name 'Name of Patient', 'Patient aadhaar number'


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
  
# IMPORT THE IRIS DATA FROM THE 
# SKLEARN MODULE
from sklearn.datasets import load_diabetes

# LOAD THE IRIS DATASET BY CALLING
# THE FUNCTION
diabetes_data = load_diabetes()
  
# PLACE THE IRIS DATA IN A PANDAS
# DATAFRAME
df = pd.DataFrame(data=diabetes_data.data, 
                  columns=diabetes_data.feature_names)
  
# DISPLAY FIRST 5 RECORDS OF THE 
# DATAFRAME



df.head()
```

Creating two colunms with null values.

```python
df['Name of Patient'] = None 
df['Patient aadhaar number'] = None 
df.head()
```

Creating random integers for aadhar card

```python
import numpy as np
random_data = np.random.randint(11111, 99999, size=len(df))
df['Patient aadhaar number'] = random_data
df.head()
```

Creating random names with null values

```python
import random

random_names = ['John Doe', 'Jane Smith', 'Michael Johnson', None, 'Emma Lee', 'William Brown']
random_name_column = [random.choice(random_names) for _ in range(len(df))]
df['Name of Patient'] = random_name_column


df.head()
```

```python
null_values = df.isnull() #checking null values
null_values
```

modilfing  the data

```python
df.loc[1, 's1'] = '9999'

# Modify row 2, column 3
df.loc[2, 'bmi'] = 100

# Modify row 5, column 11
df.loc[5, 'Name of Patient'] = 'Thilak'

# Modify row 96, column 5
df.loc[96, 's1'] = '11111'

# Modify row 1, column 1
df.loc[1, 'age'] = 999
```

```python
del df['Name of Patient']

df
```

Showing different plots for the above data set

```python

# Define figure size.
plt.figure(figsize=(20, 7))

# histogram
plt.subplot(1, 3, 1)
sns.histplot(df['age'], kde = True,color='r')
plt.title('Histogram')


# scatterplot
plt.subplot(1, 3, 2)
plt.scatter(df['age'],df['s1'],color = 'g')
plt.title('Scatterplot')
    
# barplot
plt.subplot(1, 3, 3)
sns.barplot(x = 'age', y = 'sex', data = df)   
plt.title('Barplot')
    
    
plt.show()
```
